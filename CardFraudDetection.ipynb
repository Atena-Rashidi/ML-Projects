# Unbalanced Dataset
An unbalanced dataset (or imbalanced dataset) occurs when the classes in a classification problem are not represented equally. For example, in a binary classification problem, if 90% of the data belongs to class 0 and only 10% to class 1, the dataset is unbalanced. This can lead to biased models that perform well on the majority class but poorly on the minority class.

**Solutions for Handling Unbalanced Datasets:**

**1.	Resampling Techniques:**

**- Under-Sampling:** Reduces the number of instances in the majority class to balance the dataset. This can help prevent the model from being biased towards the majority class.

**- Over-Sampling:** Increases the number of instances in the minority class by duplicating them or generating new synthetic samples (e.g., using SMOTE - Synthetic Minority Over-sampling Technique).

**2.	Algorithmic Approaches:**

**- Cost-Sensitive Learning:** Modifies the learning algorithm to pay more attention to the minority class by assigning higher misclassification costs to the minority class.

**- Ensemble Methods:** Combines multiple models to improve performance, such as using balanced random forests or boosting techniques that focus on the minority class.

**3.	Evaluation Metrics:**
- Use metrics that are more informative for unbalanced datasets, such as precision, recall, F1-score, and the area under the ROC curve (AUC-ROC), instead of accuracy.

```python
legit_samle = legit.sample(n = 492)
legit_samle.shape
```

```python
credit_card_data_balanced = pd.concat([legit_samle, fraud], axis = 0)
print(credit_card_data_balanced.shape)
credit_card_data_balanced.head()
```

```python
credit_card_data_balanced['Class'].value_counts()
```

```python
credit_card_data_balanced.groupby('Class').mean()
```

# Splitting the Data into Features and Target

```python
X = credit_card_data_balanced.drop('Class', axis = 1)
y = credit_card_data_balanced['Class']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)
print(X.shape, x_train.shape, x_test.shape) #printing the shapes of the data
```

# Model Training

```python
model = LogisticRegression(max_iter=10000)
```

```python
model.fit(x_train, y_train)
y_pred_training = model.predict(x_train)
accuracy_train = accuracy_score(y_train, y_pred_training)
print("Accuracy on Training Data:", accuracy_train)
```

```python
model.fit(x_test, y_test)
y_pred_testing = model.predict(x_test)
accuracy_test = accuracy_score(y_test, y_pred_testing)
print("Accuracy on Testing Data:", accuracy_test)
```
